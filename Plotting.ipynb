```octave
t = linspace(-2*pi, 2*pi);
sint = sin(t);

%% Set up the figure
figure(1); % open a new figure
clf;       % clear the frame
hold on;   % allow multiple changes to be made
grid on;   % make a nice grid

xlabel('xlabel')
ylabel('ylabel')

p = plot(t, sint); % plot the data as a line
set(p, 'LineWidth', 10)
```
